```python
import pandas as pd
from sklearn.metrics import confusion_matrix
```

```python
base_df = pd.read_csv('./results/0_base_data_no_result.csv')
actual_values = base_df['label']

with open('./results/t0_qwen_no_thought_results.txt', 'r') as file_reader:
    qwen_no_thought_raw = file_reader.readlines()

qwen_no_thought = [i.lower().strip() for i in qwen_no_thought_raw]

with open('./results/t0_gemma_results.txt', 'r') as file_reader:
    gemma_raw = file_reader.readlines()

gemma = [i.lower().strip() for i in gemma_raw]

with open('./results/t0_phi_results.txt', 'r') as file_reader:
    phi_raw = file_reader.readlines()

phi = [i.lower().strip() for i in phi_raw]

with open('./results/t0_llama_results.txt', 'r') as file_reader:
    llama_raw = file_reader.readlines()

llama = [i.lower().strip() for i in llama_raw]
```

```python
conf_matrix = confusion_matrix(actual_values, qwen_no_thought)
labels = [['tp', 'fn'],['fp', 'tn']]

print("CM of qwen_no_thought")
for i, _ in enumerate(labels):
    for j in zip(labels[i], conf_matrix.tolist()[i]):
        print(j)
```

```python
gem = pd.read_csv("./results/gemma-3-4b-it_multi_runs.csv")
actual_values = gem['labels']
gem.head()
```

```python
llama = pd.read_csv("./results/llama-3.2-3b-instruct_multi_runs.csv")
llama.head()
```

```python
phi = pd.read_csv("./results/phi-4-mini-instruct_multi_runs.csv")
phi.head()
```

```python
tp, fn, fp, tn = dict(), dict(), dict(), dict()
results = {'gemma': gem,
           'llama': llama,
           'phi': phi}

for result in results:

    tp[result] = list()
    fn[result] = list()
    fp[result] = list()
    tn[result] = list()

    for col_name in results[result]:

        if col_name != 'labels':

            pred = results[result][col_name]
            conf_matrix = confusion_matrix(actual_values, pred)
            cm_list = [item for sublist in conf_matrix.tolist() for item in sublist]
            tp[result].append(cm_list[0])
            fn[result].append(cm_list[1])
            fp[result].append(cm_list[2])
            tn[result].append(cm_list[3])


    



# for j, i in enumerate(gem):
    
#     # print(i)
#     if i != 'labels':
        
#         pred = gem[i]
#         conf_matrix = confusion_matrix(actual_values, pred)
#         labels = [['tp', 'fn'],['fp', 'tn']]

#         print(f"CM of Phi iteration {i}")
#         for i, _ in enumerate(labels):
#             for j in zip(labels[i], conf_matrix.tolist()[i]):
#                 print(j)
        
#         break
```

```python
import numpy as np

names = ['gemma', 'llama', 'phi']
data = [np.array(tp[n]) for n in names]

stats = {}
for name in names:
    arr = data[names.index(name)]
    stats[name] = {
        'min':    int(arr.min()),
        'median': float(np.median(arr)),
        'max':    int(arr.max()),
        'std':    float(arr.std(ddof=1))
    }

fig, ax = plt.subplots(figsize=(8, 6))

bp = ax.boxplot(
    data,
    labels=names,
    patch_artist=True,
    boxprops=dict(facecolor='#AEDFF7', edgecolor='#0D3B66'),
    medianprops=dict(color='#EE964B', linewidth=2),
    whiskerprops=dict(color='#0D3B66', linewidth=1.5),
    capprops=dict(color='#0D3B66', linewidth=1.5),
    flierprops=dict(marker='o', markerfacecolor='#D1495B', alpha=0.6)
)

ax.set_ylabel("True Positives")
ax.set_title("Plot with all models true positives with stats")

# Get existing numeric ticks:
existing = list(ax.get_yticks())

max_to_name = {stats[n]['max']: n for n in names}

for mx in max_to_name:
    if mx not in existing:
        existing.append(mx)

custom_ticks = sorted(existing)
labels_for_ticks = []
for t in custom_ticks:
    if t in max_to_name:
        labels_for_ticks.append(f"{max_to_name[t]}: {t}")
    else:
        labels_for_ticks.append(str(int(t)))

ax.set_yticks(custom_ticks)
ax.set_yticklabels(labels_for_ticks)

stats_text = ""
for n in names:
    s = stats[n]
    stats_text += (
        f"{n}:\n"
        f"  min = {s['min']}\n"
        f"  median = {s['median']:.1f}\n"
        f"  max = {s['max']}\n"
        f"  std = {s['std']:.3f}\n\n"
    )

box_props = dict(boxstyle='round', facecolor='white', edgecolor='black', alpha=0.8)
ax.text(
    1.05, 0.5, stats_text,
    transform=ax.transAxes,
    fontsize=10,
    verticalalignment='center',
    bbox=box_props
)

plt.tight_layout(rect=[0, 0, 0.75, 1.0])
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


s1 = tp['gemma']
s2 = tp['llama']
s3 = tp['phi']

arr1 = np.array(s1)
arr2 = np.array(s2)
arr3 = np.array(s3)

global_min = min(arr1.min(), arr2.min(), arr3.min())
global_max = max(arr1.max(), arr2.max(), arr3.max())
bins = np.linspace(global_min, global_max, 12)

stats = {}
for label, arr in [("gemma", arr1), ("llama", arr2), ("phi", arr3)]:
    stats[label] = {
        "min":   arr.min(),
        "max":   arr.max(),
        "median":float(np.median(arr)),
        "mean": float(arr.mean()),
        "std":  float(arr.std(ddof=1))
    }

colors = {
    "gemma": "#0D3B66",  
    "llama": "#EE964B",  
    "phi": "#D1495B",  
}

plt.figure(figsize=(8, 5))

plt.hist(arr1,
         bins=bins,
         alpha=0.5,
         color=colors["gemma"],
         edgecolor="black",
         label="Run 1 (hist)",
        )

plt.hist(arr2,
         bins=bins,
         alpha=0.5,
         color=colors["llama"],
         edgecolor="black",
         label="Run 2 (hist)",
        )

plt.hist(arr3,
         bins=bins,
         alpha=0.5,
         color=colors["phi"],
         edgecolor="black",
         label="Run 3 (hist)",
        )

linestyles = {
    "min_max":           "--",
    "median":            "-",
    "mean":              "-.",
    "mean±std":          ":",
}

for label, arr in [("gemma", arr1), ("llama", arr2), ("phi", arr3)]:
    c = colors[label]
    mn = stats[label]["min"]
    mx = stats[label]["max"]
    med = stats[label]["median"]
    mean = stats[label]["mean"]
    std = stats[label]["std"]

    plt.axvline(mn,
                color=c,
                linestyle=linestyles["min_max"],
                linewidth=1,
                label=f"{label} min ={mn}")
    plt.axvline(mx,
                color=c,
                linestyle=linestyles["min_max"],
                linewidth=1,
                label=f"{label} max ={mx}")

    plt.axvline(med,
                color=c,
                linestyle=linestyles["median"],
                linewidth=2,
                label=f"{label} median ={med:.1f}")

    plt.axvline(mean,
                color=c,
                linestyle=linestyles["mean"],
                linewidth=2,
                label=f"{label} mean ={mean:.1f}")

    plt.axvline(mean + std,
                color=c,
                linestyle=linestyles["mean±std"],
                linewidth=1,
                label=f"{label} mean+1σ ={mean+std:.1f}")
    plt.axvline(mean - std,
                color=c,
                linestyle=linestyles["mean±std"],
                linewidth=1,
                label=f"{label} mean−1σ ={mean-std:.1f}")

plt.xlabel("Reading value")
plt.ylabel("Count")
plt.title("Histograms of 3 Runs with Min / Max / Median / Mean / ±1σ")
plt.xlim(global_min - 50, global_max + 50)   # some padding
plt.ylim(0, None)  # auto‐scale y‐axis
plt.grid(axis="y", linestyle=":", alpha=0.5)

plt.legend(loc="upper left", bbox_to_anchor=(1.01, 1), fontsize="small", ncol=1, framealpha=0.8)
plt.tight_layout(rect=[0, 0, 0.75, 1.0])  # leave room on right for legend
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

arr = np.array(tp['gemma'])
mn   = arr.min()
mx   = arr.max()
med  = np.median(arr)
mean = arr.mean()
std  = arr.std(ddof=1)

plt.figure(figsize=(3, 5))

plt.errorbar(x=0,
             y=med,
             yerr=[[med - mn], [mx - med]],
             fmt="o",
             color="#0D3B66",
             ecolor="#AEDFF7",
             elinewidth=10,
             label=f"Median ={med:.1f}\n(min→max)"
            )

plt.errorbar(x=0.3,
             y=mean,
             yerr=std,
             fmt="s",
             color="#D1495B",
             ecolor="#D1495B",
             capsize=5,
             label=f"Mean ={mean:.1f}\n±1σ ={std:.1f}"
            )

plt.xlim(-1, 1)
plt.ylim(0, mx * 1.05)
plt.xticks([], []) 
plt.ylabel("Reading value")
plt.title("Summary stats: median ± [min,max], mean ± std")
plt.legend(loc="upper right", bbox_to_anchor=(1.3, 1))
plt.tight_layout()
plt.show()
```
